# 1. Datasets, DataLoaders, DataBunch

```python
# we will use normalization used in other models
normalization = ([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
mean, std = normalization
target_type = ['attr']
relevant_attribute_names = ['Smiling', 'Young', 'Male', 'Eyeglasses', 'Attractive']
batch_size = 8
path_to_celeba = '/mnt/datadrive/asya/datasets/celeba'


my_transforms = transforms.Compose(
    [transforms.Resize((224,224)), 
     transforms.ToTensor(), 
     transforms.Normalize(mean, std)]) #imagenet transforms
celeba_train_ds = CelebADataset(path_to_celeba, 
                       split='train', 
                       target_type=target_type, 
                       relevant_attribute_names=relevant_attribute_names, 
                       transform=my_transforms)
celeba_valid_ds = CelebADataset(path_to_celeba, 
                       split='valid', 
                       target_type=target_type, 
                       relevant_attribute_names=relevant_attribute_names, 
                       transform=my_transforms)
celeba_test_ds = CelebADataset(path_to_celeba, 
                       split='test', 
                       target_type=target_type, 
                       relevant_attribute_names=relevant_attribute_names, 
                       transform=my_transforms)
train_loader = DataLoader(celeba_train_ds, batch_size=batch_size, shuffle=True)
valid_loader = DataLoader(celeba_valid_ds, batch_size=batch_size, shuffle=True)
test_loader = DataLoader(celeba_test_ds, batch_size=batch_size, shuffle=True)

fastai_databunch = DataBunch(train_dl=train_loader, valid_dl=valid_loader, test_dl=test_loader)
```

# 3. Architecture

## 3.1. Define the model

Here, we will start with a base architecture and then create separate heads on top of it which will be learned. The architecture in itself will also be fine-tuned later on little by little. The architecture will have multiple outputs, one for each attribute.

```python
from torch.nn import BCEWithLogitsLoss

class InterfaceResNet(nn.Module):
    # here, we want our model to have multi-task losses for several attributes. 
    # we will do like in the paper, that is classify:

    # 1. - smile/non_smile
    # 2. - young/old
    # 3. - male/female
    # 4. - glasses/not glasses
    # 5. - attractive
    
    # !!! I will not calculate yaw pose. not really sure how to and plus it is not so interesting.
    
    def __init__(self, arch, ps=0.5):
        # ps: dropout multiplier
        super(InterfaceResNet,self).__init__()
        self.encoder = create_body(arch)
        
        # create a fully connected layer for each of the targets. we assume that all of them contain one
        # output layer
        self.fc1 = create_head(4096, 1, ps=ps)
        self.fc2 = create_head(4096, 1, ps=ps)
        self.fc3 = create_head(4096, 1, ps=ps)
        self.fc4 = create_head(4096, 1, ps=ps)
        self.fc5 = create_head(4096, 1, ps=ps)
               
    def forward(self, x):
        x = self.encoder(x)
        smile = self.fc1(x)
        age = self.fc2(x)
        gender = self.fc3(x)
        glasses = self.fc4(x)
        attractive = self.fc5(x)
        
        return [smile, age, gender, glasses, attractive]
```

## 3.2 Define the Loss and Metrics

TODO: Test with a learnable parameter for the loss, such that instead of just taking the sum of each subloss, we multiply it with some variable which the model learns.

```python
class MultiTaskLoss(nn.Module):
    def __init__(self):
        super(MultiTaskLoss, self).__init__()
        self.my_multipliers = nn.Parameter(torch.zeros((5)))
        
    def forward(self, preds, attributes):
        bce_entropy = BCEWithLogitsLoss()
        batch_size, nb_att = attributes.shape
        full_loss = 0
        for attribute_num in range(len(preds)):
            pred_for_attribute = preds[attribute_num]#[0]
            target_for_attribute = attributes[:, attribute_num].unsqueeze(1)
            loss_for_attribute = bce_entropy(pred_for_attribute, target_for_attribute)
            precision = torch.exp(-self.my_multipliers[attribute_num])
            loss = precision * loss_for_attribute + self.my_multipliers[attribute_num]
            full_loss += loss
        return full_loss
```

Let's find the best hyperparameters for the model

```python
hyperparameters = {'wd': np.logspace(1, -3, 5),
                   'dropout': np.linspace(0.5, 1., 6)}
```

```python
# weight decay
hp = 'wd'
all_losses = []
all_lrs = []
model = InterfaceResNet(arch=models.resnet50, ps=0.8)
for val in hyperparameters[hp]:
    learn = Learner(fastai_databunch, model, loss_func=loss, wd=val, callback_fns=ShowGraph, metrics=accuracies)
    learn.split([learn.model.encoder[:6],
                     learn.model.encoder[6:],
                     nn.ModuleList([learn.model.fc1, 
                                    learn.model.fc2, 
                                    learn.model.fc3, 
                                    learn.model.fc4, 
                                    learn.model.fc5])]) 
    
    learn.freeze()
    learn.lr_find(start_lr=1e-7, end_lr=1., num_it=100)
    learn.recorder.plot(suggestion=True)
    
    # get the losses and associated learning rates
    losses = np.copy(np.array(learn.recorder.losses))
    lrs = np.copy(np.array(learn.recorder.lrs))
    
    all_losses.append(losses)
    all_lrs.append(lrs)
```

```python
#minnum = 40
minnum = 40
maxnum = 90
lr = all_lrs[0][minnum:maxnum]

fig, ax = plt.subplots(1,1)
#         ax.plot(lrs, losses)
#         ax.set_ylabel("Loss")
#         ax.set_xlabel("Learning Rate")
#         ax.set_xscale('log')
        
        
for i, loss in enumerate(all_losses):
    
    cur_legend = "wd - " + str(hyperparameters[hp][i])
    print(cur_legend)
    
    cur_loss = loss[minnum:maxnum]
    best_grad = (np.gradient(cur_loss)).argmin()
    ax.plot(lr, loss[minnum:maxnum], label=cur_legend)
    ax.scatter(lr[best_grad], loss[best_grad + minnum])
    ax.set_xscale('log')
    #plt.legend(cur_legend)
plt.legend()  
```
