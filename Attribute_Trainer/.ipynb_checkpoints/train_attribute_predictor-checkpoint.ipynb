### 1.2 Create dataloader from pytorch custom dataset

First, let's make the dataframe for the train/validation set

```python
# construct DataFrame with labels

df = pd.DataFrame()
df['names'] = celeba_ds.filenames

# get the labels

tiled = np.tile(celeba_ds.relevant_attribute_names, reps=(df.shape[0], 1))
mask = np.array(celeba_ds.relevant_attributes == 0) # because the mask below signifies INvalid entries
attr = np.ma.masked_array(tiled, mask=mask)
list_atts = attr.tolist()
labels = [' '.join([aa for aa in a if aa is not None]) for a in list_atts]
df['labels'] = labels

# validation set
df['is_valid'] = celeba_ds.splits == 1 # if it is a validation entity, since we are only with train/validation data
```

# 2. Training

### 2.1 Create the metrics we want to visualize
We will create separate accuracies for each of the separate attributes and then a general accuracy which treats everything together the same.
For the separate accuracies for each class, this requires us to know the class number, which can be accessed like this:

```python
celeba.vocab

# Attractive : 0
# Eyeglasses:  1
# Male:        2
# Smiling:     3
# Young:       4
```

```python
# create new custom metrics which is the accuracy of all of the classes separately


def abstract_accuracy(preds, attributes, attribute_num):
    pred_for_attribute = (preds[:, attribute_num].sigmoid()) >= 0.5 #my threshold is 0.5 here
    target_for_attribute = attributes[:, attribute_num].bool()
    return (pred_for_attribute == target_for_attribute).float().mean() 

# obtain all the metrics, this will be a separate accuracy for each attribute, and then a multi-accuracy
metrics = []
for i in range(len(celeba.vocab)):
    current_vocab = celeba.vocab[i]
    func_i = partial(abstract_accuracy, attribute_num=i)
    metrics.append(func_i)
metrics.append(accuracy_multi) # for the multi accuracy
```

### 2.2 Create the learner
Here, we will use a pretrained resnet50 (on ImageNet) for out base network which we will fine-tune.

```python
checkpoint_path = '/home/datadrive/asya/checkpoints/celeba_attributes/attribute_predictor_40' # this is for the 5 attributes
#checkpoint_path = '/mnt/datadrive/asya/checkpoints/celeba_attributes/attribute_predictor_all'
```

```python
learn = cnn_learner(celeba, resnet50, metrics=metrics)
```

We will freeze the body and only train the head for 2 epochs, then we will train the entire model for 5 epochs. 

### 2.3 Train, save, load model
We will train the model by first freezing the first part of the resnet50 model up to the last layer, training like this for 2 epochs, with a learning rate of 0.023 (according to our learning rate finder). Then, we will unfreeze the rest of the model and train the model with discriminative learning rates (varying from 0.0023 to 0.023 throughout the model) for 5 epochs. The total time this takes is about 1:15 on our Quadro GV100 32gb GPU.

```python
# load the model if we already have it
learn.load(checkpoint_path)
learn.metrics = metrics

should_train = False
```

```python
if should_train:
    r = learn.lr_find()
    print(r)
```

```python
if should_train:
    s = time.time()
    learn.fine_tune(epochs=5, base_lr=2.3e-2, freeze_epochs=2) 
    e = time.time()
    print("total time to train:", e - s)

    # save model !
    learn.save(checkpoint_path)
```

```python
if should_train:
    # try to get better results
    learn.fit(5)
```

```python
if should_train:
    # load the model, it did not get better
    learn.load(checkpoint_path)
    learn.metrics = metrics
```

### 2.4 Validate on validation and test sets

```python
def print_metric_results(results):
    print("loss on valdiation set:", results[0])
    print()
    print("Attribute Name : Accuracy (%)")
    max_string_length = max([len(a) for a in learn.dls.vocab])
    for i in range(1, len(learn.dls.vocab) + 1):
        vocab_name = learn.dls.vocab[i - 1] + (max_string_length - len(learn.dls.vocab[i - 1]) + 1) * ' '
        print("\t{} :\t {}".format(vocab_name, results[i]))
    print("Multi Accuracy : {}".format(results[-1]))
```

```python
# results on validation set
validation_res = learn.validate()

print_metric_results(validation_res)
```

```python
# results on test set
test_res = learn.validate(dl=celeba_test.train)

print_metric_results(test_res)
```

As we can see, we have very good results, both for test and validation sets!!!

# 3. Testing on wild images

We will test on unaligned normal images and also aligned images to see if aligning images is worth it.

```python
# Get the image files in the test directories
unaligned_images = Path("/home/data/images/test_images")
aligned_images = Path("/home/data/images/aligned_images")

unaligned_images = get_image_files(unaligned_images)
aligned_images = get_image_files(aligned_images) # this is with fastai2.data.transforms
```

```python
print(aligned_images)
```

```python
def predict_for_images(learn, images):
    for image in images:
        im_name = image.name.lower()

        print("image:", im_name)
        labels, preds, pred_scores = learn.predict(image)
        #print("prediction:", pred)
        for i in range(len(celeba.vocab)):
            pred = preds[i]
            pred_score = pred_scores[i]
            label = celeba.vocab[i]
            if not pred:
                s = 'NOT '
            else:
                s = ''
            print(s + label + ' - ' + str(round(pred_score.item(), 3)))

        plt.imshow(Image.open(image))
        plt.show()
```

```python
predict_for_images(learn, unaligned_images)
```

As we can see, many results are wrong and many should be obvious results are not confident (eyeglasses...) <br>
Let's see if predicting on aligned images gives better results?

```python
predict_for_images(learn, aligned_images)
```
